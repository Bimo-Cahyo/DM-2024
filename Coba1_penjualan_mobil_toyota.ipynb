```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = 'penjualan_mobil_toyota.csv'  # Update this to your actual file path
data = pd.read_csv(file_path)
```

```python
# Convert 'Bulan' column to datetime format
data['Bulan'] = pd.to_datetime(data['Bulan'])
```

```python
# Set 'Bulan' as the index for easier time-series analysis
data.set_index('Bulan', inplace=True)
```

```python
# Display the first few rows to confirm changes
display(data.head())
```

```python
# Plot the time series data
plt.figure(figsize=(14, 7))
plt.plot(data.index, data['Penjualan'], marker='o', linestyle='-', color='b')
plt.title('Toyota Car Sales Over Time')
plt.xlabel('Year-Month')
plt.ylabel('Number of Cars Sold')
plt.grid(True)
plt.show()
```

```python
# Moving Average to smooth out trends (optional)
data['Moving_Avg'] = data['Penjualan'].rolling(window=3).mean()
```

```python
# Plot with Moving Average
plt.figure(figsize=(14, 7))
plt.plot(data.index, data['Penjualan'], marker='o', color='b', label='Monthly Sales')
plt.plot(data.index, data['Moving_Avg'], color='r', linestyle='-', linewidth=2, label='3-Month Moving Average')
plt.title('Toyota Car Sales with Moving Average')
plt.xlabel('Year-Month')
plt.ylabel('Number of Cars Sold')
plt.legend()
plt.grid(True)
```

```python
# Seasonal Decomposition (Optional)
from statsmodels.tsa.seasonal import seasonal_decompose

decomposition = seasonal_decompose(data['Penjualan'], model='additive', period=12)
decomposition.plot()
plt.show()
```
